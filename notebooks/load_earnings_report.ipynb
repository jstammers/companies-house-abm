```python
url = "http://download.companieshouse.gov.uk/historicmonthlyaccountsdata.html"
```

```python
import requests
from bs4 import beautifulsoup
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
links = soup.find_all("a")
```

```python
# get total size
total_size = 0
for link in links:
    if "zip" in link.text:
        value = link.text.split(" ")[1][1:-1]
        print(value)
        if "Mb" in value:
            total_size += float(value[:-2])
        elif "Kb" in value:
            total_size += float(value[:-2]) / 1000
        elif "Gb" in value:
            total_size += float(value[:-2]) * 1000
        else:
            continue
```
